## knn 구현

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris.keys()
```

```python
iris.target_names
```

```python
iris.feature_names
```

```python
iris.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = iris.data
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 5)

#neighbor의 수 = 5
```

```python
knn.fit(X_train, y_train)

#훈련
```

```python
from sklearn import metrics
```

```python
print("Accuracy : ", knn.score(X_test, y_test))
```

## K 구하기

```python
k_range = range(1,50) # k범위 1~ 50
scores = {}
scores_list = []

for k in k_range :
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train, y_train)
    y_prediction = knn.predict(X_test)
    scores[k] = metrics.accuracy_score(y_test, y_prediction)
    scores_list.append(metrics.accuracy_score(y_test, y_prediction))
    
    # K의 지정해 놓은 범위 내에서 어떤 K값이 가장 정확도가 높은지를 찾고자 함
```

```python
plt.plot(k_range, scores_list)
plt.xlabel('K')
plt.ylabel('Accuracy')
```

k가 약 10~20 일때 정확도 가장 높다.
